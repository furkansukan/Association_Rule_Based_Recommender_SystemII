```python
############################################
# 1. Veri Ön İşleme
############################################

# !pip install mlxtend
import pandas as pd
pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
pd.set_option('display.width', 500)
# çıktının tek bir satırda olmasını sağlar.
pd.set_option('display.expand_frame_repr', False)
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Görev 1: Veriyi Hazırlama
# Adım 1: csv dosyasını okutunuz.

df_ = pd.read_csv("/kaggle/input/retail-data-set/file_out2.csv")
df = df_.copy()
df.head()
```

```python
df.drop("Unnamed: 0", axis=1, inplace = True)
df.dtypes
df.head()
```

```python
df.isnull().sum()
```

```python
# Görev 2: Birliktelik Kuralları Üretiniz ve Öneride bulununuz

# Adım 1: Aşağıdaki gibi sepet, hizmet pivot table’i oluşturunuz.

pivot_df = df.pivot_table(index="InvoiceID", columns="ProductID", aggfunc= lambda x: 1, fill_value=0)
pivot_df.head()
```

```python
pivot_df.shape
```

```python
pivot_df = pivot_df.astype("bool")
pivot_df.head()
```

```python
# Adım 2: Birliktelik kurallarını oluşturunuz.

frequent_itemsets = apriori(pivot_df,
                            min_support=0.05,
                           use_colnames=True)
```

```python
frequent_itemsets
```

```python
frequent_itemsets.sort_values("support", ascending=False)
```

```python
rules = association_rules(frequent_itemsets, metric="support", min_threshold=0.01)
rules
```

```python
ProductID = 192

sorted_rules = rules.sort_values("lift", ascending=False)

recommendation_list = []

for i, product in enumerate(sorted_rules["antecedents"]):
    for j in list(product):
        if j == ProductID:
            recommendation_list.append(list(sorted_rules.iloc[i]["consequents"])[0])
            
recommendation_list
```
